### Train a naive model on Tournament Matchups


```python
df_tourgames = pd.read_csv('data/Tournament Matchups.csv')
df_tourgames.describe()
```

```python
df_tourgames.dropna(inplace=True)
df_tourgames.describe()
```

```python
df_tourgames.head(n=20)
```

```python
#dataframe transformation
df_team1 = df_tourgames.iloc[::2].reset_index(drop=True)
df_team2 = df_tourgames.iloc[1::2].reset_index(drop=True)
df_team1.head()
```

```python
df_games = pd.DataFrame({
    'YEAR': df_team1['YEAR'],
    'ROUND': df_team1['ROUND'],
    'CURRENT ROUND': df_team1['CURRENT ROUND'],
    'TEAM NO 1': df_team1['TEAM NO'],
    'TEAM 1': df_team1['TEAM'],
    'SEED 1': df_team1['SEED'],
    'SCORE 1': df_team1['SCORE'],
    'TEAM NO 2': df_team2['TEAM NO'],
    'TEAM 2': df_team2['TEAM'],
    'SEED 2': df_team2['SEED'],
    'SCORE 2': df_team2['SCORE'],
    'TEAM 1 WIN': (df_team1['SCORE'] > df_team2['SCORE']).astype(int)
})

df_games.head()
```

### First actual model using Logistic Regression. Gary: Dunk number might be able to quantify morale

```python
df_enhanced2 = df_games.merge(
    bartovikANdf[['YEAR', 'TEAM NO', 'BADJ O', 'BADJ D', 'WIN%']],
    left_on=['YEAR', 'TEAM NO 1'],
    right_on=['YEAR', 'TEAM NO'],
    how='left'
)
df_enhanced2.rename(columns={'BADJ O': 'BADJ O 1', 'BADJ D': 'BADJ D 1', 'WIN%':'WIN% 1'}, inplace=True)
df_enhanced2.head()
```

```python
df_enhanced2.drop('TEAM NO', axis=1, inplace=True)

df_enhanced2 = df_enhanced2.merge(
    bartovikANdf[['YEAR', 'TEAM NO', 'BADJ O', 'BADJ D', 'WIN%']],
    left_on=['YEAR', 'TEAM NO 2'],
    right_on=['YEAR', 'TEAM NO'],
    how='left'
)
df_enhanced2.rename(columns={'BADJ O': 'BADJ O 2', 'BADJ D': 'BADJ D 2', 'WIN%':'WIN% 2'}, inplace=True)
df_enhanced2.drop('TEAM NO', axis=1, inplace=True)

df_enhanced2[df_enhanced2['YEAR'] == 2024].head(10)
```

```python
df_enhanced2.describe()
```

```python
df_enhanced2['BADJ O DIFF'] = df_enhanced2['BADJ O 1'] - df_enhanced2['BADJ O 2']
df_enhanced2['BADJ D DIFF'] = df_enhanced2['BADJ D 1'] - df_enhanced2['BADJ D 2']
df_enhanced2['WIN DIFF'] = df_enhanced2['WIN% 1'] - df_enhanced2['WIN% 2']

df_enhanced2.head()
```

```python
df_enhanced2['SEED DIFF'] = df_enhanced2['SEED 1'] - df_enhanced2['SEED 2']
df_enhanced2.head()
```

```python
import seaborn as sns
df_enhanced2MAP = df_enhanced2[['SEED DIFF', 'BADJ O DIFF', 'BADJ D DIFF', 'WIN DIFF', 'TEAM 1 WIN']]
correlation_matrix = df_enhanced2MAP.corr()
plt.figure(figsize=(8, 6)) # Adjust figure size for readability
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, MinMaxScaler
from sklearn.pipeline import Pipeline
df_enhanced2train = df_enhanced2[df_enhanced2['YEAR'] < 2023]
df_enhanced2test = df_enhanced2[df_enhanced2['YEAR'] >= 2023]
##X = df_filtered_sessions[['sim2instructor', 'sim2case','sim2book', 'ln_length','instance',
##                          'interaction', 'top_speaker']]
##y = df_filtered_sessions['ESL_val']
##X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=8)
X_train = df_enhanced2train[['BADJ O DIFF', 'BADJ D DIFF', 'SEED DIFF']]
X_test = df_enhanced2test[['BADJ O DIFF', 'BADJ D DIFF', 'SEED DIFF']]
y_train = df_enhanced2train['TEAM 1 WIN']
y_test = df_enhanced2test['TEAM 1 WIN']

scalar = StandardScaler()
X_train_processed = scalar.fit_transform(X_train)
X_test_processed = scalar.transform(X_test)
X_train_processed
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
Cs= [0.1, 1, 10, 100] 
solvers= ['lbfgs', 'liblinear', 'sag', 'saga']
for solver in solvers:
    for C in Cs:
        logreg_model = LogisticRegression(random_state=1, max_iter=1000, solver=solver, C=C)
        logreg_model.fit(X_train_processed, y_train)
        y_pred = logreg_model.predict(X_test_processed)
        print(f"Solver: {solver}, C: {C}")
        print(classification_report(y_test, y_pred, digits=4))
```

```python
###now we use the knn model

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

ks = [3, 5, 7, 9, 11, 15]
for k in ks:
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train_processed, y_train)
    y_pred = knn_model.predict(X_test_processed)
    print(f"Results for k={k}")
    print(classification_report(y_test, y_pred, digits=4))
```

```python
###SVC
Cs = [0.1, 0.2, 0.5, 0.8, 1, 1.2, 1.5, 2, 5, 10]
solvers = ['linear', 'rbf', 'poly', 'sigmoid']
for solver in solvers:
    for C in Cs:
        print(f"Results for solver={solver}, C={C}")
        svc_model = SVC(C=C, kernel=solver)
        svc_model.fit(X_train_processed, y_train)
        y_pred_svc = svc_model.predict(X_test_processed)
        print(classification_report(y_test, y_pred_svc, digits=4))
```

```python
random_forest_model = RandomForestClassifier(random_state=8)
random_forest_model.fit(X_train_processed, y_train)
y_pred_rf = random_forest_model.predict(X_test_processed)
print(classification_report(y_test, y_pred_rf, digits=4))
```
